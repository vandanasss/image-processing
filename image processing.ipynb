```python
import numpy as np
import scipy as sp
import itertools as it
import matplotlib.pyplot as plt
from scipy import signal
from matplotlib.animation import FuncAnimation
from matplotlib.image import imread
```

```python
import cv2
```

```python
img=plt.imread("retina.jpg")
```

```python
img
```

```python
fig,ax=plt.subplots(figsize=(10,10))
ax.imshow(img)
ax.axis('off')
plt.show()
```

```python
# Sharpen Image
kernel_sharpening = np.array([[-1,-1,-1], 
                              [-1,9,-1], 
                              [-1,-1,-1]])

sharpened = cv2.filter2D(img, -1, kernel_sharpening)

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(sharpened)
ax.axis('off')
ax.set_title('Sharpened Image')
plt.show()
```

```python
# edge Image
kernel_edge = np.array([[-1,-1,-1], 
                              [-1,8,-1], 
                              [-1,-1,-1]])

edge = cv2.filter2D(img, -1, kernel_edge)

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(edge)
ax.axis('off')
ax.set_title('edge Image')
plt.show()
```

```python
img_gray = cv2.cvtColor(sharpened, cv2.COLOR_RGB2GRAY)   #gray  image of sharpened image
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(img_gray, cmap='Greys')
ax.axis('off')
ax.set_title('Grey Image of sharpened image')
plt.show()
```

```python
img_gray = cv2.cvtColor(edge, cv2.COLOR_RGB2GRAY)    #gray image of edge filter
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(img_gray, cmap='Greys')
ax.axis('off')
ax.set_title('Grey Image of edge image')
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)    #gray image of  given image
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(img_gray, cmap='Greys')
ax.axis('off')
ax.set_title('Grey Image')
plt.show()
```
